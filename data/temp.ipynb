```python
# open train.csv as pandas dataframe
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

train = pd.read_csv('train.csv')
valid = pd.read_csv('dev.csv')
test = pd.read_csv('test.csv')
```

```python
train.shape, valid.shape, test.shape
```

```python
# make a new column called 'label', 'binary-label' in test dataframe and set it to -1
test['label'] = -1
test['binary-label'] = -1
test.head()
```

```python
# concatenate train and valid
data = pd.concat([train, valid, test], axis=0)
```

```python
data.shape
```

```python
data.head()
```

```python
# find alphabet in 'sentence_1' and 'sentence_2' in train df
data_include_alphabet = data[data['sentence_1'].str.contains('[a-zA-Z]') | data['sentence_2'].str.contains('[a-zA-Z]')]
data_include_alphabet
```

```python
data_include_sharp = data[data['sentence_1'].str.contains('#') | data['sentence_2'].str.contains('#')]
data_include_sharp
```

```python
# find html tag in 'sentence_1' and 'sentence_2' in train df
# use regex to find html tag
# except <PERSON> tag
import re
data_include_html = data[data['sentence_1'].str.contains('<[^P].*>') | data['sentence_2'].str.contains('<[^P].*>')]
data_include_html.head(100)
```

```python
# find U+200B in 'sentence_1' and 'sentence_2' in train df
data_include_U200B = data[data['sentence_1'].str.contains('\u200b') | data['sentence_2'].str.contains('\u200b')]
data_include_U200B
```

```python
# find U+FEFF in 'sentence_1' and 'sentence_2' in train df
data_include_UFEFF = data[data['sentence_1'].str.contains('\ufeff') | data['sentence_2'].str.contains('\ufeff')]
data_include_UFEFF
```

```python
# find () in 'sentence_1' and 'sentence_2' in train df
# data_include_parenthesis = data[data['sentence_1'].str.contains('\(\)')]
data_include_parenthesis = data[data['sentence_1'].str.contains('\(') | data['sentence_2'].str.contains('\(')]
data_include_parenthesis
```

```python
# find sequence of space in 'sentence_1' and 'sentence_2' in train df
data_include_space = data[data['sentence_1'].str.contains('  ') | data['sentence_2'].str.contains('  ')]
data_include_space
```

```python
# find emoji in 'sentence_1' and 'sentence_2' in train df
data_include_emoji = data[data['sentence_1'].str.contains('[\U0001F600-\U0001F64F]') | data['sentence_2'].str.contains('[\U0001F600-\U0001F64F]')]
data_include_emoji
```
