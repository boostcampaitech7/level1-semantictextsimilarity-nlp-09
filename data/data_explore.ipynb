```python
import re

def find_non_korean_characters(sentence):
    return re.findall(r'[^가-힣a-zA-Z0-9\s]', sentence)

non_korean = train['sentence_1'].apply(find_non_korean_characters)

non_korean_list = non_korean.tolist()
punctuations = []
for i in non_korean_list:
    for j in i:
        if j not in punctuations:
            punctuations.append(j)
punctuations
```
